```python
def absolute(x):
    if x >= 0:
        return x
    else:
        return -x
```

```python
print("the absolute of -5 is: ", absolute(-5))
print("the absolute of 10 is: ", absolute(10))
```

        <h3>ساختار شرطی match-case</h3>
        <h4>(مرور جلسه قبل)</h4>
        <ul>
            <li>دستور match-case برای تطبیق الگوها (Pattern Matching) استفاده می‌شود.</li>
            <li>مقدار یک متغیر یا عبارت با چندین حالت (case) مقایسه می‌شود.</li>
            <li>اگر یکی از حالت‌ها مطابقت داشته باشد، کد مربوط به آن اجرا می‌شود.</li>
            <li>حالت پیش‌فرض با علامت _ مشخص می‌شود (یعنی هر چیزی که با هیچ case دیگری تطبیق نکند).</li>
            <li>میتوان از تمامی اپراتور های منطقی نیز درون caseها استفاده کرد.</li>
        </ul>
        <h4>ویژگی‌ها و مزایا:</h4>
        <ul>
            <li>خوانایی بیشتر نسبت به if-elif-else.</li>
            <li>وجود حالت پیش‌فرض ( _ case) برای پوشش همه‌ی شرایط دیگر.</li>
            <li>مناسب برای کدهایی که نیاز به بررسی چندین حالت دارند.</li>
        </ul>
    </div>
</body>
</html>

```python
day = 3

match day:
    case 1:
        print("شنبه")
    case 2:
        print("یکشنبه")
    case 3:
        print("دوشنبه")
    case 4:
        print("سه‌شنبه")
    case _:
        print("روز نامعتبر")
```

        در این مسائل معمولاُ جمله های اول مشخص هستند و بقیه جمله ها زنجیر وار به یکدیگر وابسته اند.<br>
        اکنون با توجه به دانش خود تابع فیبوناچی را در پایتون پیاده کنید.<br>

```python
def fibonachi(n):
    if n == 0:
        return 1
    elif n == 1:
        return 1
    else:
        return fibonachi(n-1)+ fibonachi(n-2)
```

        <h3>ویژگی ها (Attributes):</h3>
            در پایتون، Attribute (ویژگی) داده یا خصوصیتی است که به یک شیء (Object) یا کلاس (Class) تعلق دارد. Attribute می‌تواند اطلاعاتی مثل نام، سن یا هر داده‌ی دیگر را نگه دارد.
            به بیانی ساده، متغیرهایی که به یک کلاس یا یک شی انتساب داده می‌شود صفت یا ویژگی یا Attribute خوانده می‌شوند.
            هر شیء می‌تواند چندین Attribute داشته باشد:
        <h4>1- ویژگی شئ (Instance Attribute):</h4>
مقدار این دسته از ویژگی ها (Arrtibutes) به ازای هر شی منحصر‌به‌فرد است. برای نمونه صفت‌هایی همچون نام، نام‌خانوادگی، سن و جنسیت برای هر یک از اشیای کلاس «شخص» قابل تعریف است.
این ویژگی ها در متد جادویی __init__ تعریف میشود.


```python
class Person:
    def __init__(self, name, age):
        self.name = name
        self.age = age
    def greet(self):
        print("hello! I'm ", self.name, "and I have ", self.age, "years old")
```

```python
taha = Person("taha" , 25)
taha.greet()
```

        <h4>2- ویژگی کلاس (Class Attribute):</h4>
به Attributeهای خاص یک کلاس گفته می‌شود و در واقع متغیرهایی است که درون کلاس و خارج از متدها تعریف می‌گردند. کاربرد این Attributeها به اشتراک گذاشتن یک یا چند مقدار یکسان در بین تمام اشیاست.
تمام اشیای یک کلاس به Class Attributeهای آن کلاس دسترسی دارند:

```python
class Person:
    number_of_leg = 2
    numbet_of_hand = 2
    
    def __init__(self, name, age):
        self.name = name
        self.age = age
    
    def greet(self):
        print("hello ", self.name)

    def info(self):
        print("all humans have ", self.number_of_leg, "legs and ", self.numbet_of_hand, "hands")
```

        <h3>ارث‌بری (Inheritance):</h3>
ارث‌بری (Inheritance) راهکاری برای ساخت یک کلاس جدید برای استفاده از جزئیات موجود در یک کلاس بدون ویرایش آن است. کلاس جدیدی که شکل می‌گیرد، یک کلاس مشتق شده (یا کلاس فرزند) است. به طور مشابه، کلاس موجود، کلاس مبنا (کلاس والد) است.<br>
برای تعریف ارث بری فقط کافیست در تعریف کلاس فرزند، نام کلاس والد را درون پرانتز جلوی نام کلاس فرزند بنویسیم.


```python
class Student(Person):
    def __init__(self, name, age, student_id):
        super().__init__(name, age)
        self.student_id = student_id
```

```python
taha = Student("taha" , 25, "402621xxxx")
taha.greet()
```

        <h1>قالب‌بندی رشته‌ها (String Formatting):</h1>
        در پایتون برای قالب‌بندی رشته‌ها (String Formatting) روش‌های مختلفی وجود دارد. این قابلیت به ما اجازه می‌دهد مقادیر متغیرها را داخل رشته قرار دهیم یا خروجی را با شکل و ظاهر دلخواه نمایش دهیم.
        <h3>انواع Format String در پایتون:</h3>
        <h4>۱. استفاده از عملگر % (روش قدیمی)</h4>
         در ابتدا در پایتون از عملگر % برای قالب‌بندی رشته‌ها استفاده می‌شد. این روش شبیه به قالب‌بندی رشته‌ها در زبان C ست و هنوز هم در پایتون پشتیبانی می‌شود، هرچند دیگر توصیه نمی‌شود چون خوانایی و انعطاف کمتری دارد.<br>
برای استفاده در رشته، از جای نگهدار هایی (placeholders) %s, %d, %f استفاده می‌کنیم.
بعد از رشته، با عملگر % مقادیر جایگزین را قرار می‌دهیم.

```python
name = "Taha"
age = 25
text = "My name is %s and I am %d years old." % (name, age)
print(text)
```

        <h4> ۲. متد ()str.format (روش کلاسیک‌تر)</h4>
        بعد از روش قدیمی %، پایتون متد ()str.format را معرفی کرد تا قالب‌بندی رشته‌ها انعطاف‌پذیرتر و خواناتر شود. این متد امکان استفاده از جای‌نگهدارها (placeholders) داخل رشته را می‌دهد که با مقادیر دلخواه جایگزین می‌شوند.<br>

<h5>نحوه استفاده:</h5>
در رشته، جای‌نگهدارها با {} مشخص می‌شوند.
سپس با فراخوانی ()format. روی رشته، مقادیر جایگزین می‌شوند.

```python
name = "Taha"
age = 25
text = "My name is {} and I am {} years old.".format(name, age)
print(text)
```

        <h4>۳. f-string (روش مدرن و سریع)</h4>
       از نسخه Python 3.6 به بعد، یک روش بسیار ساده و قدرتمند برای قالب‌بندی رشته‌ها معرفی شد که به آن f-string یا formatted string literal می‌گویند. این روش هم سریع‌تر است و هم خواناتر، چون می‌توانیم مستقیماً متغیرها و حتی عبارات پایتون را داخل رشته قرار دهیم.

 داخل {} می‌توانیم نام متغیر یا هر عبارت پایتون را قرار دهی.

```python
name = "Taha"
age = 25
text = f"My name is {name} and I am {age} years old."
print(text)
```

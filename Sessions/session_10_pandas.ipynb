```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
```

<!DOCTYPE html>
<html lang="fa", dir="rtl">
<body >
    <div style="direction: rtl; unicode-bidi: embed;">
        <h2>۱. ساخت DataFrame (جدول داده)</h2>
        <p>مهم‌ترین ساختار داده در پانداس، <b>DataFrame</b> است که شبیه یک شیت اکسل دارای سطر و ستون است.</p>
    </div>
</body>
</html>

```python
# ساخت دستی یک جدول نمرات
data = {
    'Name': ['Ali', 'Sara', 'Reza', 'Maryam', 'Omid'],
    'Age': [20, 21, 19, 22, 20],
    'Grade': [18.5, 19.0, 14.5, 17.0, 11.0],
    'City': ['Tehran', 'Isfahan', 'Tehran', 'Shiraz', 'Isfahan']
}

df = pd.DataFrame(data)

# نمایش جدول
df
```

<!DOCTYPE html>
<html lang="fa", dir="rtl">
<body >
    <div style="direction: rtl; unicode-bidi: embed;">
        <h2>۲. اطلاعات کلی و آمار (Info & Describe)</h2>
        <p>چطور بفهمیم داده‌های ما چه خصوصیاتی دارند؟</p>
    </div>
</body>
</html>

```python
# اطلاعات ساختاری (تعداد سطر، نوع داده‌ها، داده‌های خالی)
print("--- Info ---")
df.info()

# آمار توصیفی (میانگین، انحراف معیار، مینیمم، ماکزیمم)
print("\n--- Statistics ---")
print(df.describe())
```

<!DOCTYPE html>
<html lang="fa", dir="rtl">
<body >
    <div style="direction: rtl; unicode-bidi: embed;">
        <h2>۳. انتخاب و فیلتر کردن (Filtering)</h2>
        <p>انتخاب ستون خاص یا سطرهایی که شرط خاصی دارند.</p>
    </div>
</body>
</html>

```python
# انتخاب یک ستون
grades = df['Grade']
print(f"All grades: {grades.values}")

# فیلتر: دانشجویانی که نمره بالای 17 دارند
top_students = df[df['Grade'] > 17]
print("\nTop Students (>17):")
print(top_students)
```

<!DOCTYPE html>
<html lang="fa", dir="rtl">
<body >
    <div style="direction: rtl; unicode-bidi: embed;">
        <h2>۴. گروه‌بندی (GroupBy)</h2>
        <p>مشابه Pivot Table در اکسل. مثلاً میانگین نمرات بر اساس شهر محل سکونت.</p>
    </div>
</body>
</html>

```python
# میانگین نمره به تفکیک شهر
city_stats = df.groupby('City')['Grade'].mean()

print("Average Grade by City:")
print(city_stats)
```

<!DOCTYPE html>
<html lang="fa", dir="rtl">
<body >
    <div style="direction: rtl; unicode-bidi: embed;">
        <h2>۵. رسم نمودار سریع</h2>
        <p>پانداس مستقیماً با Matplotlib متصل است و می‌توانید به راحتی نمودار بکشید.</p>
    </div>
</body>
</html>

```python
df.plot(kind='bar', x='Name', y='Grade', color='skyblue', title='Student Grades')
plt.plot([0, 4], [12, 12], 'r--', label='Pass Limit') # خط قرمز قبولی
plt.legend()
plt.show()
```

        <h3>ساختار شرطی match-case</h3>
        <ul>
            <li>دستور match-case برای تطبیق الگوها (Pattern Matching) استفاده می‌شود.</li>
            <li>مقدار یک متغیر یا عبارت با چندین حالت (case) مقایسه می‌شود.</li>
            <li>اگر یکی از حالت‌ها مطابقت داشته باشد، کد مربوط به آن اجرا می‌شود.</li>
            <li>حالت پیش‌فرض با علامت _ مشخص می‌شود (یعنی هر چیزی که با هیچ case دیگری تطبیق نکند).</li>
            <li>میتوان از تمامی اپراتور های منطقی نیز درون caseها استفاده کرد.</li>
        </ul>
    </div>
</body>
</html>

```python
day = 3

match day:
    case 1:
        print("شنبه")
    case 2:
        print("یکشنبه")
    case 3:
        print("دوشنبه")
    case 4:
        print("سه‌شنبه")
    case _:
        print("روز نامعتبر")
```

```python
value = "apple"

match value:
    case "apple" | "banana":
        print("میوه")
    case "carrot":
        print("سبزیجات")
    case _: # other cases
        print("نامشخص")
```

        <h1>توابع بازگشتی</h1>
        در برخی مسائل به الگویی بازگشتی بر میخوریم. برای مثال دنباله فیبوناچی:<br>
        <p style="font-family: 'Times New Roman', Times, serif; font-size: 16px; font-style: italic;">
            F(n) = F(n-1) + F(n-2)<br>
            F(0) = 1<br>
            F(1) = 1
        </p>
        در این مسائل معمولاُ جمله های اول مشخص هستند و بقیه جمله ها زنجیر وار به یکدیگر وابسته اند.<br>
        اکنون با توجه به دانش خود تابع فیبوناچی را در پایتون پیاده کنید.<br>
        راهنمایی: میتوانید تابع را درون خود فراخوانی کنید
    </div>
</body>
</html>

```python
def fibonachi(n):
    if n == 0:
        return 1
    elif n == 1:
        return 1
    else:
        return fibonachi(n-1)+ fibonachi(n-2)
```

        <h3>استفاده از with (روش بهتر):</h3>
        همانطور که در کد ها دقت کردید بعد از هر بار باز کردن فایل ها بایستی آن را ببندیم.<br>
        روش بهتر و معمول تر کار با فایل ها استفاده از with است که فایل را بعد از استفاده به صورت خودکار میبندد.
    </ul>

```python
with open("data.txt", "r") as f:
    content = f.read()
    print(content)
```

        <h1>شی‌گرایی در پایتون</h1>
        «زبان برنامه‌نویسی پایتون» (Python Programming Language)، یک زبان برنامه‌نویسی چند پارادایمی است. بدین معنا که از رویکردهای برنامه‌نویسی گوناگونی پشتیبانی می‌کند. یک راهکار محبوب برای حل مسائل برنامه‌نویسی، ساخت «شی» (Object) است.<br>
        زبان‌های برنامه‌نویسی که از این رویکرد و در واقع پارادایم برنامه‌نویسی پشتیبانی می‌کنند را «شی‌گرا» (Object Oriented) و برنامه‌نویسی با این پارادایم را «برنامه‌نویسی شی‌گرا» (Object Oriented Programming) می‌گویند.  یک شی، دو مشخصه دارد:
        <ul>
            <li>خصیصه (Attribute)</li>
            <li>رفتار (Method)</li>
        </ul>
        مفهوم OOP در پایتون، بر ساخت کدهای قابل استفاده مجدد تمرکز دارد. این مفهوم با عنوان «DRY» (سرنامی برای عبارت Don't Repeat Yourself) نیز  شناخته شده است و بدان معنا است که کاری که یک‌بار انجام شده را نباید مجدد تکرار کرد و در واقع، می‌توان از کدها استفاده مجدد کرد. استفاده مجدد از کدهای نوشته شده، یکی از مباحث مهم در بحث مهندسی نرم‌افزار است. در پایتون، مفهوم OOP از چند اصل پایه‌ای پیروی می‌کند که در ادامه شرح داده شده‌اند.
        <ul>
            <li>ارث‌بری (Inheritance): فرایند استفاده از جزئیات از کلاس جدید بدون ویرایش کلاس موجود را ارث‌بری گویند.</li>
            <li>کپسوله‌سازی (Encapsulation): پنهان‌سازی جزئیات خصوصی یک کلاس از اشیای دیگر را کپسوله‌سازی گویند.</li>
            <li>چندریختی (Polymorphism): مفهوم استفاده از عملیات متداول به روش‌های متفاوت برای ورودی‌های داده متفاوت را چندریختی یا پلی‌مورفیسم گویند.</li>
            <li>انتزاعی (Abstraction): فرآیندی است که در پی آن جزئیات پیاده‌سازی و عملکرد برنامه پنهان می‌شود و فقط می‌توان به اطلاعات درخواستی و اساسی دسترسی داشت.</li>
        </ul>

        <h2>توضیح جامع‌تر ژنراتورها</h2>
        هنگامی که تابع ژنراتور فراخوانی می‌شود در اصل یک شی ژنراتور (که رفتاری بسیار شبیه iterator ها دارد) تولید می‌شود.
        <br>
        <ul>
            <li>پس می‌توان با همان تابع <code>next()</code> مقادیر آن را یکی یکی خواند.</li>
            <li>همچنین می‌توان با توابع <code>list()</code>، <code>tuple()</code>، <code>set()</code> مقادیر ژنراتور را درون یک لیست یا ست یا تاپل ذخیره کرد.</li>
            <li>همچنین می‌توان از آن درون ساختار <code>for</code> به عنوان شی قابل پیمایش استفاده کرد.</li>
        </ul>
    </div>
</body>

```python
def generator_func(n):
    for i in range(n):
        yield i
```

```python
gen_1 = generator_func(10)
print(next(gen_1))
print(list(gen_1)) # عنصر اول از بین رفت
```

```python
gen_2 = generator_func(12)
print(type(gen_2))
```

```python
for i in generator_func(15):
    print(i, end="\t")
```

        مزیت ژنراتور ها نسبت به تابع های معمولی در سرعت فوق العاده بالای آنها و همچنین مصرف کمتر منابع سیستم (مموری) است.

```python
import time
import sys

def normal_func(n):
    return [x for x in range(n)]

def generator_func(n):
    for i in range(n):
        yield i

time_1 = time.time()
first_list = normal_func(100_000_000) 
print(f"time to create list: {time.time()-time_1:.10f} second")
print(f"memory usage for first_list:\n{sys.getsizeof(first_list)} B\n{sys.getsizeof(first_list)/1024:.3f} KB\n{sys.getsizeof(first_list)/(1024*1024):.3f} MB")
print("-"*50)

del first_list
print("memory free up!")
print("-"*50)

time_2 = time.time()
second_list = generator_func(100_000_000)
print(f"time to create generator: {time.time()-time_2:.10f} second")
print(f"memory usage for second_list:\n{sys.getsizeof(second_list)} B\n{sys.getsizeof(second_list)/1024:.3f} KB\n{sys.getsizeof(second_list)/(1024*1024):.3f} MB")
print("-"*50)
```

        <h2>۲. آرگومان‌های متغیر (*args و **kwargs)</h2>
        <p>گاهی اوقات نمی‌دانیم کاربر قرار است چند ورودی به تابع بدهد. مثلاً تابعی که میانگین نمرات را حساب می‌کند، شاید برای ۵ نمره صدا زده شود و شاید برای ۵۰ نمره.</p>
        <h3>*args (آرگومان‌های پوزیشنی نامحدود)</h3>
        <ul>
            <li>ورودی‌ها را به صورت یک <b>تاپل (Tuple)</b> جمع‌آوری می‌کند.</li>
            <li>نام <code>args</code> قراردادی است، مهم علامت ستاره <code>*</code> است.</li>
        </ul>
    </div>
</body>
</html>

```python
def calculate_average(*grades):
    print(f"Received data: {grades} (Type: {type(grades)})")
    if not grades:
        return 0
    return sum(grades) / len(grades)

print("Avg 1:", calculate_average(18, 19, 20))
print("Avg 2:", calculate_average(10, 15, 12, 18, 20))
```

        <h3>**kwargs (آرگومان‌های کلیدی نامحدود)</h3>
        <ul>
            <li>ورودی‌ها را به صورت یک <b>دیکشنری (Dictionary)</b> جمع‌آوری می‌کند (نام=مقدار).</li>
            <li>نام <code>kwargs</code> مخفف Keyword Arguments است. مهم دو ستاره <code>**</code> است.</li>
            <li>برای تنظیمات (Config) یا ویژگی‌های متغیر یک شیء بسیار مفید است.</li>
        </ul>
    </div>
</body>
</html>

```python
def print_student_info(name, **details):
    print(f"Student: {name}")
    for key, value in details.items():
        print(f"  - {key}: {value}")
    print("-" * 20)

print_student_info("Ali", age=20, city="Tehran")
print_student_info("Sara", grade=19.5, major="Electrical Eng", year=1)
```
